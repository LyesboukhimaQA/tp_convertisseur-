### A4) Mini‑analyse YAML (bonus +2 pts)  
Considérez cet extrait :

```yaml
name: Python Tests
on: [push, pull_request]

jobs:
  test:
    runs-on: ubuntu-latest
    steps:
      - name: Cloner le dépôt
        uses: actions/checkout@v3
      - name: Installer Python
        uses: actions/setup-python@v4
        with:
          python-version: '3.10'
      - name: Installer les dépendances
        run: pip install -r requirements.txt
      - name: Lancer les tests
        run: pytest tests/
```

**B1.** À quoi sert `with: python-version: '3.10'` ? *(1 pt)*
with: python-version: '3.10' sert à spécifier la version de Python (3.10) qui sera installée et utilisée sur le runner pour exécuter le job
**B2.** À quoi sert `with: uses: actions/checkout@v3` ? *(1 pt)* 
uses: actions/checkout@v3 sert à cloner le dépôt GitHub dans la machine virtuelle (runner) afin que le code soit disponible pour les étapes suivantes (installation, tests, etc.).
